### Loading the DataSet

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('synthetic_student_dataset.csv')  # <-- Replace with your actual filename

# Display the first 5 rows to check the data
print("First 5 rows of the dataset:")
print(df.head())

# Display column names to ensure they match
print("\nColumn names:")
print(df.columns)
```

```python
print("Summary Statistics:")
print(df.describe())
```

### Correlation Matrix (HeatMap)

```python
features = ['comprehension', 'attention', 'focus', 'retention', 'engagement_time']
target = 'assessment_score'

# Compute correlation matrix
corr_matrix = df[features + [target]].corr()

# Print correlation matrix
print("Correlation Matrix:")
print(corr_matrix)

# Plot heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Cognitive Skills and Assessment Score')
plt.show()
```

### PairPlot Visualization 

```python
sns.pairplot(df[features + [target]])
plt.suptitle("Pairplot of Cognitive Skills and Assessment Score", y=1.02)
plt.show()
```
